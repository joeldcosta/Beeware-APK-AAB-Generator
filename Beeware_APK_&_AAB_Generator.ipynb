#How to install Beeware on your Computer

- Creating Environment:-
```
###############################
create a folder -> beeware-tutorial
cmd -> cd beeware-tutorial
cmd -> py -m venv beeware-venv | python -m venv beeware-venv
close cmd
Go back to beeware-tutorial folder
create a batch file
batch -> start beeware-venv\Scripts\activate
run
###############################
```

- Install briefcase
```
###############################
open batch file
python -m pip install briefcase
############(if once installed no need to install it again.) 
Creating a new app:-
batch -> briefcase new (this will generate new application) 
```
Fill following data refer :- https://docs.beeware.org/en/latest/tutorial/tutorial-1.html
```
batch -> cd helloworld
batch -> briefcase dev (this will start the app)
edit app.py according to what you like
batch -> briefcase dev (run again to check)
################################
```

- Now Creating APK commands:-
```
################################
batch -> briefcase create android (this downloads an Android app template and adds your Python code to it.)
batch -> briefcase build android (create into APK)
batch -> briefcase run android (install in your device)
################################
```
- if you get errors then install NDK 
- add **local.properties** to app folder
```
i.e 
sdk.dir=C:\\Users\\user\\AppData\\Local\\Android\\Sdk
ndk.dir=C:\\Users\\user\\AppData\\Local\\Android\\Sdk\\ndk\\21.3.6528147
```

#Create a file with name - "local.properties"
```

ndk.dir=C:\\android-sdk\\ndk-bundle
```

# Get AAB from APK
Source:- https://stackoverflow.com/questions/53040047/generate-apk-file-from-aab-file-android-app-bundle
#Download bundletool.jar

````
!wget https://github.com/google/bundletool/releases/download/1.7.1/bundletool-all-1.7.1.jar
````

#Rename bundletool

````
!mv bundletool-all-1.7.1.jar bundletool.jar
````

````
!java -jar bundletool.jar build-apks \
    --bundle=final.aab \
    --output=final.apks \
    --mode=universal \
    --ks=keystores/beeware.keystore \
    --ks-pass=pass:123456789 \
    --ks-key-alias=beeware \
    --key-pass=pass:123456789
!unzip final.apks
````
